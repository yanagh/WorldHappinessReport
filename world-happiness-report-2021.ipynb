# Analysis

```python
#Total Ladder Score grouped by Regional Indicator, rounded to 1 decimal point
dfri = df.groupby(["Regional indicator"]).sum().sort_values("Ladder score", ascending=False).head(20) 
dfri = dfri[["Ladder score"]].round(1) 
dfri.reset_index(inplace=True) 
dfri
```

```python
#lets have a closer look at Sub-Saharan Africa
dfssa=df[df['Regional indicator']=='Sub-Saharan Africa']
dfssa.head()
```

```python
#lets have a closer look at Western Europe
dfwe=df[df['Regional indicator']=='Western Europe']
dfwe.head()
```
